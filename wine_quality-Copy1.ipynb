```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sb
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
wine=pd.read_csv('WineQT.csv')
```

```python
y=wine['quality']
x=wine[['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides',
      'free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']]
```

```python
sb.histplot(wine['quality'])
```

```python
co_matrix=wine[['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides',
      'free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']].corr()
sb.heatmap(co_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sb.pairplot(wine,x_vars=['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol'],y_vars='quality',kind='scatter')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
winelr=LinearRegression()
```

```python
winelr.fit(xtrain,ytrain)
```

```python
winelr.coef_
```

```python
pd.DataFrame(winelr.coef_,index=x.columns,columns=['mycoef'])
```

```python
pr=winelr.predict(xtest)
```

```python
r2_score(ytest,pr)
```

```python
mean_absolute_error(ytest,pr)
```

```python
mean_squared_error(ytest,pr)
```
